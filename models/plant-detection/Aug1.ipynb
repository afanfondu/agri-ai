```python
import os
import cv2
import numpy as np

# Define the paths for original and augmented images
original_data_dir = r"C:\Users\rajdi\Desktop\IMP Projects\Ayurvedic Medicinal Plant Identification\Original_Data"
augmented_data_dir = r"C:\Users\rajdi\Desktop\IMP Projects\Ayurvedic Medicinal Plant Identification\Augmented_Data"

# Create the augmented data directory if it doesn't exist
if not os.path.exists(augmented_data_dir):
    os.makedirs(augmented_data_dir)

# Function to perform data augmentation
def augment_image(image):
    # Apply your augmentation techniques here
    # # For example, you can apply rotations, flips, scaling, etc.
    augmented_images = []
    augmented_images.append(image) # Original image
    # Add three more augmented versions of the image
    # # Example augmentation:
    augmented_images.append(cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE))
    augmented_images.append(cv2.flip(image, 1)) # Horizontal flip
    augmented_images.append(cv2.flip(image, 0)) # Vertical flip
    return augmented_images

# Loop through class folders
for class_name in os.listdir(original_data_dir):
    class_dir = os.path.join(original_data_dir, class_name)
    if os.path.isdir(class_dir):
        # Create a corresponding class folder in the augmented data directory
        augmented_class_dir = os.path.join(augmented_data_dir, class_name)
        if not os.path.exists(augmented_class_dir):
            os.makedirs (augmented_class_dir)
        
        # Loop through leaf images in the class folder
        for leaf_image_name in os.listdir(class_dir):
            leaf_image_path = os.path.join(class_dir, leaf_image_name)
            if os.path.isfile(leaf_image_path):
                # Load the original image
                original_image = cv2.imread(leaf_image_path)

                # Perform data augmentation
                augmented_images = augment_image(original_image)

                # Save augmented images in the corresponding class folder
                for idx, augmented_image in enumerate (augmented_images):
                    augmented_image_name = f"{leaf_image_name.split('.')[0]}_{idx}.jpg"
                    augmented_image_path = os.path.join(augmented_class_dir, augmented_image_name)
                    cv2.imwrite(augmented_image_path, augmented_image)
```
